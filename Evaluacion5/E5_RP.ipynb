# **Ejercicio 1 - Clasificación (C=2.5)**

Implemente y compare los siguientes algoritmos de clasificación para la base de datos Breast Cancer Wisconsin. Para el entrenamiento de los modelos, use el subconjunto ```x_train, y_train```. Una vez entrenados, compare el desempeño de los diferentes clasificadores dibujando las fronteras de decisión para el subconjunto de datos ```x_test, y_test```. Al final, deberá mostrar un resultado similar aunque no necesariamente igual al de la figura de la parte inferior.

1. Regresión logística
2. Clasificador kNN

**NOTA:** Los datos tomados de Breast Cancer Wisconsin se generarán de acuerdo a su número se CC. Por favor, ingrese su número en la línea indicada.

[Info del dataset](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29)



````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #PARA REGRESION LOGISTICA
from sklearn.neighbors import KNeighborsClassifier as kNN #PARA KNN
from sklearn.datasets import load_breast_cancer # Base de datos

np.random.seed( 1140415092 ) ##--> Ingrese aquí su número de CC
idx = np.random.randint(0,30,2)

data = load_breast_cancer()
y = data.target[:] 
x = data.data[:, idx]; x[np.where(y==1),:] -= .05;

print ("###---- Info del dataset ----###")
print ("x, y shapes: {0}, {1}".format(x.shape, y.shape))
print ("Nro muestras: {0}".format(x.shape[0]))
print ("Nro caracteristicas: {0}".format(x.shape[1]))
print ("Nro clases: {0}".format(len(np.unique(y))))
print ("Nombres de las clases: {0}".format(list(data.target_names)))
print ("ID de las clases: {0}\n\n".format(np.unique(y)))


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .10)
# Mostrar la base de datos que debera procesar
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, edgecolors='k', marker="x", cmap=plt.cm.Paired, label="train")
plt.scatter(x_test[:, 0] , x_test[:, 1] , c=y_test , edgecolors='k', marker="o", cmap=plt.cm.Paired, label="test")
plt.legend(), plt.xticks([]), plt.yticks([])
````

# ENTRENAMIENTO DE REGRESION LOGISTICA

````python
# Ingrese el código para realizar la clasificación con los algoritmos aquí. 
regression = LogisticRegression()
regression.fit (x, y)

# Utilice cuantas celdas de código necesite
x_min, x_max = x[:, 0].min(), x[:, 0].max()
y_min, y_max = x[:, 1].min(), x[:, 1].max()

x_grid, y_grid = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Predecir los valores del grid
prediction = regression.predict(np.c_[x_grid.ravel(), y_grid.ravel()])
print (prediction.shape)

````

````python
# Muestre las fronteras de decisión de cada algoritmo. Obtenga un gráfico similar al siguiente
prediction = prediction.reshape(x_grid.shape)

# Dibujar el grid y las fronteras de decision
_, ax = plt.subplots (figsize=(8,5))
ax.pcolormesh(x_grid, y_grid, prediction, cmap = plt.cm.Paired)

# Imprimir las muestras con el scatter
ax.scatter (x[:, 0], x[:, 1],  c=y, edgecolors='k', cmap=plt.cm.Paired)
ax.set_xlim(x_grid.min(), x_grid.max())
ax.set_ylim(y_grid.min(), y_grid.max())
ax.set_xticks(())
ax.set_yticks(())
````

# ENTRENAMIENTO DE KNN NEIGHBOURS

````python
neighbors = 2 #VECINOS A IDENTIFICAR
knn = kNN(neighbors)
knn.fit (x, y)

x_min, x_max = x[:, 0].min(), x[:, 0].max()
y_min, y_max = x[:, 1].min(), x[:, 1].max()

#FRONTERAS DE DECISION
x_grid, y_grid = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
#PREDICCION
prediction = knn.predict(np.c_[x_grid.ravel(), y_grid.ravel()])
print (prediction.shape)

#ENCAJE
prediction = prediction.reshape(x_grid.shape)

# GRAFICAR EL ENCAJE EN LAS FRONTERAS
_, ax = plt.subplots (figsize=(8,5))
ax.pcolormesh(x_grid, y_grid, prediction, cmap = plt.cm.Paired)

ax.scatter (x[:, 0], x[:, 1],  c=y, edgecolors='k', cmap=plt.cm.Paired)
ax.set_xlim(x_grid.min(), x_grid.max())
ax.set_ylim(y_grid.min(), y_grid.max())
ax.set_xticks(())
ax.set_yticks(())
````

# **Ejercicio 2 - Clasificación (C=2.5)**

Implemente y compare los siguientes algoritmos de clasificación para la base de datos. Para el entrenamiento de los modelos, use el subconjunto ```x_train, y_train```. Una vez entrenados, compare el desempeño de los diferentes clasificadores dibujando las fronteras de decisión para el subconjunto de datos ```x_test, y_test```. Al final, deberá mostrar un resultado similar aunque no necesariamente igual al de la figura de la parte inferior.

1. Regresión logística
2. Clasificador kNN

**NOTA:** Los datos para la base de datos se generarán de acuerdo a su número se CC. Por favor, ingrese su número en la línea indicada.



````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.model_selection import train_test_split


np.random.seed( 1140415092 ) ##--> Ingrese aquí su número de CC

xMa, yMa = make_classification(n_samples = 1000, n_features=2, n_classes = 3, n_redundant=0, n_informative=2, n_clusters_per_class=1)

x_trainMa, x_testMa, y_trainMa, y_testMa = train_test_split(xMa, yMa, test_size = .10)
# Mostrar la base de datos que debera procesar
plt.scatter(x_trainMa[:, 0], x_trainMa[:, 1], c=y_trainMa, edgecolors='k', marker="x", cmap=plt.cm.Paired, label="train")
plt.scatter(x_testMa[:, 0] , x_testMa[:, 1] , c=y_testMa , edgecolors='k', marker="o", cmap=plt.cm.Paired, label="test")
plt.legend(), plt.xticks([]), plt.yticks([])


print ("###---- Info del dataset ----###")
print ("Nro muestras: {0}".format(xMa.shape[0]))
print ("Nro caracteristicas: {0}".format(xMa.shape[1]))
print ("Nro clases: {0}\n\n".format(len(np.unique(yMa))))

print ("###------ Datos de entrenamiento ------###")
print ("Nro muestras: {0}".format(x_trainMa.shape[0]))
print ("Nro caracteristicas: {0}".format(x_trainMa.shape[1]))
print ("ID, #muestras por clases: {0}\n\n".format(np.unique(y_trainMa, return_counts=True)))

print ("###------ Datos de prueba ------###")
print ("Nro muestras: {0}".format(x_testMa.shape[0]))
print ("Nro caracteristicas: {0}".format(x_testMa.shape[1]))
print ("ID, #muestras por clases: {0}\n\n".format(np.unique(y_testMa, return_counts=True)))
````

# REGRESION LOGISTICA

````python
regressionMa = LogisticRegression()
regressionMa.fit(xMa, yMa)
# Utilice cuantas celdas de código necesite

x_minMa, x_maxMa = xMa[:, 0].min(), xMa[:, 0].max()
y_minMa, y_maxMa = xMa[:, 1].min(), xMa[:, 1].max()

x_gridMa, y_gridMa = np.meshgrid(np.arange(x_minMa, x_maxMa, 0.01), np.arange(y_minMa, y_maxMa, 0.01))

# Predecir los valores del grid
predictionsMa = regressionMa.predict(np.c_[x_gridMa.ravel(), y_gridMa.ravel()])
print (predictionsMa.shape)

````

````python
predictionsMa = predictionsMa.reshape(x_gridMa.shape)

_, axMa = plt.subplots (figsize=(8,5))
axMa.pcolormesh(x_gridMa, y_gridMa, predictionsMa, cmap = plt.cm.Paired)

axMa.scatter (x[:, 0], x[:, 1],  c=y, edgecolors='k', cmap=plt.cm.Paired)
axMa.set_xlim(x_gridMa.min(), x_gridMa.max())
axMa.set_ylim(y_gridMa.min(), y_gridMa.max())
````

# ENTRENAMIENTO KNN NEIGHBOURS

````python
neighborsM = 2
knnM = knn(neighborsM)
knnM.fit(xMa,yMa)

x_minMa, x_maxMa = xMa[:, 0].min(), xMa[:, 0].max()
y_minMa, y_maxMa = xMa[:, 1].min(), xMa[:, 1].max()

x_gridMa, y_gridMa = np.meshgrid(np.arange(x_minMa, x_maxMa, 0.01), np.arange(y_minMa, y_maxMa, 0.01))
#PREDICCION
predictionsMa = knnM.predict(np.c_[x_gridMa.ravel(), y_gridMa.ravel()])
print (predictionsMa.shape)

predictionsMa = predictionsMa.reshape(x_gridMa.shape)

# GRAFICAR EL ENCAJE EN LAS FRONTERAS
_, axMa = plt.subplots (figsize=(8,5))
axMa.pcolormesh(x_gridMa, y_gridMa, predictionsMa, cmap = plt.cm.Paired)

axMa.scatter (xMa[:, 0], xMa[:, 1],  c=yMa, edgecolors='k', cmap=plt.cm.Paired)
axMa.set_xlim(x_gridMa.min(), x_gridMa.max())
axMa.set_ylim(y_gridMa.min(), y_gridMa.max())
````
