## **Ejercicio 1**

Utilice la imagen "[figures_1.jpg](https://drive.google.com/file/d/18V8DsnqhxjZAs3alX0hO3xHfUBT5JV2I/view?usp=sharing)" para detectar todos los tipos posibles de figuras contenidas en ella (triángulos, rectángulos, pentágonos, hexágonos, heptágonos, octágonos y círculos). Subrayar en la imágen principal los contornos de todas las figuras geométricas contenidas en ella. A continuación, imprima la imagen original y al lado, la imagen con los contornos de las figuras dibujadas. Además, imprima la cantidad aproximada de cada tipo de figura geométrica contenida en la imagen original y muestrelos.

```
# Importar librerias. Incluya acá todas las librerías que requiera
import cv2
import numpy as np
from skimage.measure import find_contours, approximate_polygon, subdivide_polygon
import matplotlib.pyplot as plt
```

```
fig1 = cv2.imread("/content/drive/MyDrive/Patrones/repositorio/patrones/Evaluacion3/Images/figures_1.jpg")
fig1_gris = cv2.cvtColor(fig1, cv2.COLOR_BGR2GRAY)

#---- Introduzca su codigo aqui. Puede utilizar cuantas celdas de código considere necesarias
_, threshold = cv2.threshold(fig1_gris, 230,255,cv2.THRESH_BINARY_INV)

contours = find_contours(threshold)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,4))
ax1.imshow(fig1, cmap='gray')
ax2.imshow(threshold, cmap='gray')

for n, contour in enumerate(contours):
    ax3.plot(contour[:, 1], contour[:, 0], linewidth=3)

    # Approximate all contours in the image 
for n, contour in enumerate(contours):
    coords = approximate_polygon(contour, tolerance = 5 ) #--> Approximate
    ax2.plot(coords[:, 1], coords[:, 0], linewidth=2)

Triangles = 0
Rectangles = 0
Pentagons = 0
Hexagons = 0
Heptagons = 0
Octagons = 0
Circles = 0

for n, contour in enumerate(contours):
    coords = approximate_polygon(contour, tolerance=5)
    
    if  len(coords)-1 == 3:
     Triangles+=1
    elif 3< len(coords)-1 == 4:
      Rectangles+=1
    elif 4< len(coords)-1 == 5:
      Pentagons+=1
    elif 5< len(coords)-1 == 6:
      Hexagons+=1
    elif len(coords)-1 == 7:
      Heptagons+=1
    elif len(coords)-1 == 8:
      Octagons+=1
    elif 8< len(coords) > 20:
      Circles+=1
    else:
      0


print("triangles:", Triangles)
print("rectangles:", Rectangles)
print("Pentagons:", Pentagons)
print("Hexagons:", Hexagons)
print("Heptagons:", Heptagons)
print("Octagons:", Octagons)
print("Circles:", Circles)
```
